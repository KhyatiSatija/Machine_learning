```python
FIRST_TIME = True
data= []
COMMA, QUOTE = ',', '""'
for line in open("ghw.csv"):
    if FIRST_TIME:
        FIRST_TIME = False
    else:
        g, h, w = line.strip().split(COMMA)
        data.append((g, h, w))
```

```python
INCHES2CMS, POUNDS2KG = 2.54 , 0.453
COMMA, QUOTES = ',', '""'
FIRST_TIME = True
data = []
for line in open("ghw.csv"):
    if FIRST_TIME:
        FIRST_TIME = False
    else:
        g_raw, h_raw, w_raw = line.strip().split(COMMA)
        g = g_raw.strip(QUOTES)
        h = int(float(h_raw.strip(COMMA)) * INCHES2CMS + 0.5)
        w = int(float(w_raw.strip(COMMA)) * POUNDS2KG + 0.5)
        data.append((g, h, w))
```

```python
INCHES2CMS , POUNDS2KG = 2.54, 0.454
def inches2cms(s: str) -> int:
    return int(float(s) * INCHES2CMS + 0.5)
def pounds2kgs(s: str) -> int:
    return int(float(s) * POUNDS2KG + 0.5)
```

```python
fig, axis = plt.subplots(1, 2, figsize = (10,5))
# (1,2) means 1 row of subplots horizontally and two columns of subplots vertically.
plt.subplot(1, 2, 1) #selecting the first subplot

plt.plot(women.Height, women.Weight, "b.")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(['women'], loc = 'upper left')

plt.subplot(1, 2, 2)  #selecting the first subplot

plt.plot(men.Height, men.Weight, "k+")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(["men"], loc = 'lower right')

plt.show()
```

```python
#In the previous graphs , we observe that the x and y axis scales for both the graphs are different due to which comparison
#between women and men is not being made clearly. Hence we manually set the scales the same in both the subplots.
fig, axis = plt.subplots(1, 2, figsize = (10,5), layout='constrained')

plt.subplot(1, 2, 1) #selecting the first subplot

plt.plot(women.Height, women.Weight, "b.")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(['women'], loc = 'upper left')
plt.axis([140, 200, 30, 120])
# plt.axis([x_min, x_max, y_min, y_max])

plt.subplot(1, 2, 2)  #selecting the first subplot

plt.plot(men.Height, men.Weight, "k+")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(["men"], loc = 'lower right')
plt.axis([140, 200, 30, 120])

plt.show()
```

```python
fig, axis = plt.subplots(1, 2, figsize = (10,5), sharex = True, sharey = True)
#THis is another method to have the same scales for both the plots.
# (1,2) means 1 row of subplots horizontally and two columns of subplots vertically.
plt.subplot(1, 2, 1) #selecting the first subplot

plt.plot(women.Height, women.Weight, "b.")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(['women'], loc = 'upper left')

plt.subplot(1, 2, 2)  #selecting the first subplot

plt.plot(men.Height, men.Weight, "k+")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend(["men"], loc = 'lower right')

plt.show()
```
